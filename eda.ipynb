```python
#Searching for a specific Movie Title
df[df['Title'].isin(['Night on Earth'])]
#Can also use str.contains for finding specific words within value
```

```python
#Replacing missing values

df['Director'].fillna('Unknown', inplace=True)
df['Cast'].fillna('Unknown', inplace=True)
df['Country'].fillna('Unknown',inplace=True)

#fillna replaces missing values (NaN) with something else (such as 'Unknown')
#inplace=True means the change will be permanent in the dataset after the line of code

df.isnull().sum() #Checking how many missing values there are after modifying dataset
```

```python
#Handling duration column
df['Duration'].head(10)
```

```python
#Splitting duration into number and unit
df[['duration_value', 'duration_unit']] = df['Duration'].str.split(' ', expand=True)
df['Duration'].head(10)
```

```python
#Changing duration values to numerical format
df['duration_value'] = pd.to_numeric(df['duration_value'], errors='coerce')
df['duration_value'].head(10)
```

```python
df[['duration_value', 'duration_unit']].isnull().sum()
df['duration_unit'].head(10)
```

```python
#Handling 'Type' column

df['Type'].value_counts(dropna=False)
```

```python
df['Type'] = df['Type'].str.strip().str.title()
df['Type']
```

```python
df['Type'].value_counts()
```

```python
#Separate Movies and TV Shows

movies = df[df['Type'].str.contains('Movie', case=False, na=False)]
tv_shows = df[df['Type'].str.contains('TV Show', case=False, na=False)]

print("Number of Movies:", movies.shape[0])
print("Number of TV Shows:", tv_shows.shape[0])

```

```python
#Barchart to shows number of Movies compared to TV Shows

num_movies = movies.shape[0]
num_tv_shows = tv_shows.shape[0]

print("Number of Movies:", num_movies)
print("Number of TV Shows:", num_tv_shows)

counts = {'Movies': num_movies, 'TV Show': num_tv_shows}

plt.figure(figsize=(6,5))
sns.barplot(x=list(counts.keys()), y=list(counts.values()))
plt.title("Number of Movies VS TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")
```
